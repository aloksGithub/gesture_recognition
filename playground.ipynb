```python
from TrainNarma import *
from Models import *
import matplotlib.pyplot as plta
import numpy as np
```

```python
def createESN(**params):
    if 'units' in params:
        params['units'] = int(params['units'])
    reservoir = Reservoir(**params)
    readout = Ridge(output_dim=1)
    esn = reservoir >> readout
    return esn

def createReservoir(**params):
    if 'units' in params:
        params['units'] = int(params['units'])
    return Reservoir(**params)

pbounds = {'units': (10, 500), 'sr': (0.1, 2), 'lr': (0, 1), 'input_scaling': (-1, 1), 'rc_connectivity': (0, 1)}
optimalParams, error = optimize(pbounds, createESN, createReservoir)
```

```python
print(error)
```

```python
separabilities = []
generalizabilities = []
mcs = []
errors = []
for model in optimalParams:
    if (model['error']<4):
        separabilities.append(model['separability'])
        generalizabilities.append(model['generalizability'])
        errors.append(model['error'])
        mcs.append(model['mc'])
```

```python
# Data is 200, 50 for kr, gr
# Train data is 1, 1000

def makeGraph2D(x, y, label):
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    
    weights = np.ones_like(x) / len(x)
    ax1.hist(x, weights=weights,  color='#3480eb', edgecolor='black')
    ax2.scatter(x, y, s=10, color='red')
    ax2.set_ylabel("Error")
    ax1.set_ylabel("Probability")
    ax1.set_xlabel(label)
    
makeGraph2D(np.array(separabilities)-np.array(generalizabilities), errors, "KR-GR")
makeGraph2D(separabilities,  errors, "KR")
makeGraph2D(generalizabilities,  errors, "GR")
makeGraph2D(separabilities, generalizabilities, "KR")
# makeGraph2D(mcs, errors, "MC")
```
